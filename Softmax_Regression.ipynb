## Training

```python
X_train = np.array(df.iloc[:42000, 1:21])
m, n = np.shape(X_train)
print(m, n)
print(X_train[0,:])
X_train, X_mean, X_std = lnr.normalize(X_train)
```

```python
y_train = np.array(df.iloc[:42000, 21])
print(np.shape(y_train))
classes = np.unique(y_train)
print(classes)
num_classes = len(classes)
print(num_classes)
```

```python
w = np.random.rand(n, num_classes)
print(np.shape(w))
b = np.random.rand(num_classes)
print(np.shape(b))
```

```python
yhat_train = predict_softmax(X_train, w, b)
print(np.shape(yhat_train))
yhat_train
```

```python
y_pred_train = predict_class(X_train, w, b)
print(np.shape(y_pred_train))
y_pred_train
```

```python
cost = softmax_cost(y_train, yhat_train)
cost
```

```python
n_iter = 1000
alpha = 1.5
w, b, cost_history = softmax_gradient_descent(X_train, w, b, y_train, n_iter, alpha, classes)
```

```python
cost_history[::n_iter//50]
```

```python
yhat_train = predict_softmax(X_train, w, b)
print(np.shape(yhat_train))
y_pred_train = softmax_to_class(yhat_train)
print(np.shape(y_pred_train))
y_pred_train
```

```python
mce_train = mce(y_train, y_pred_train)
mce_train
```

```python
pd.crosstab(y_train, y_pred_train, rownames = ['Actual'], colnames = ['Predicted'])
```

```python
f1_train_macro = macro_f1_multi(y_train, y_pred_train)
f1_train_macro
```

```python
print(one_f1_multi(y_train, y_pred_train, 0))
print(one_f1_multi(y_train, y_pred_train, 1))
print(one_f1_multi(y_train, y_pred_train, 2))
print(one_f1_multi(y_train, y_pred_train, 3))
print(one_f1_multi(y_train, y_pred_train, 4))
```

```python
f1_train_micro = micro_f1_multi(y_train, y_pred_train)
f1_train_micro
```
